##Example 1: Oxygen on Platinum

We will calculate the adsorption energy of an oxygen molecule on a platinum surface. This is done by calculating the total energy for the isolated slab and for the isolated molecule. The adsorbate is then added to the slab and relaxed, and the total energy for this composite system is calculated. The adsorption energy is obtained as the sum of the isolated energies minus the energy of the composite system.

```python
from ase import Atoms
from ase.calculators.emt import EMT
from ase.constraints import FixAtoms
from ase.io import write
from ase.lattice.surface import fcc111, add_adsorbate
from ase.optimize import QuasiNewton

h = 1.90
d = 1.21

slab = fcc111('Pt', size=(4, 4, 2), vacuum=10.0)

slab.set_calculator(EMT())
e_slab = slab.get_potential_energy()

molecule = Atoms('2O', positions=[(0., 0., 0.), (0., 0., d)])
molecule.set_calculator(EMT())
e_O2 = molecule.get_potential_energy()

add_adsorbate(slab, molecule, h, 'ontop')
constraint = FixAtoms(mask=[a.symbol != 'O' for a in slab])
slab.set_constraint(constraint)
dyn = QuasiNewton(slab, trajectory='O2Pt.traj')
dyn.run(fmax=0.05)

print('Adsorption energy:', e_slab + e_O2 - slab.get_potential_energy())
write('O2Pt.png', slab * (3, 3, 1), rotation='-75x')
```

###Atoms

The Atoms object is a collection of atoms. Here is how to define a O2 molecule by directly specifying the position of two oxygen atoms:

```python
from ase import Atoms
d = 1.21
molecule = Atoms('2O', positions=[(0., 0., 0.), (0., 0., d)])
```

You can also build crystals using, for example, the lattice module which returns Atoms objects corresponding to common crystal structures. Let us make a Pt(111) surface:

```python
from ase.lattice.surface import fcc111
slab = fcc111('Pt', size=(4,4,2), vacuum=10.0)
```

###Calculators

Many ase.calculators can be used with ASE, including emt, Asap, Dacapo, GPAW, Abinit, Vasp. See the ASE home page for the full list.

In this overview we use the effective medium theory (EMT) calculator, as it is very fast and hence useful for getting started.

We can attach a calculator to the previously created Atoms objects:

```python
from ase.calculators.emt import EMT
slab.set_calculator(EMT())
molecule.set_calculator(EMT())
```

and use it to calculate the total energies for the systems by using the get_potential_energy() method from the Atoms class:

```python
e_slab = slab.get_potential_energy()
e_O2 = molecule.get_potential_energy()
```

###Structure relaxation

Let’s use the QuasiNewton minimizer to optimize the structure of the O2 molecule adsorbed on the Pt surface. First add the adsorbate to the Pt slab, for example in the on-top position:

```python
h = 1.90
add_adsorbate(slab, molecule, h, 'ontop')
```

In order to speed up the relaxation, let us keep the Pt atoms fixed in the slab by using FixAtoms from the constraints module. Only the O2 molecule is then allowed to relax to the equilibrium structure:

```python
from ase.constraints import FixAtoms
constraint = FixAtoms(mask=[a.symbol != 'O' for a in slab])
slab.set_constraint(constraint)
```

Now attach the QuasiNewton minimizer to the system and save the trajectory file. Run the minimizer with the convergence criteria that the force on all atoms should be less than some fmax:

```python
from ase.optimize import QuasiNewton
dyn = QuasiNewton(slab, trajectory='O2Pt.traj')
dyn.run(fmax=0.05)
```

###Input-output

Writing the atomic positions to a file is done with the write() function:

```python
from ase.io import write
write('O2Pt.xyz', slab)
write('O2Pt.png', slab * (3, 3, 1), rotation='-75x')
```

Reading from a file is done like this:

```python
from ase.io import read
slab_from_file = read('O2Pt.xyz')
```

If the file contains several configurations, the default behavior of the write() function is to return the last configuration. However, we can load a specific configuration by doing:

```python
read('O2Pt.traj')      # last configuration
read('O2Pt.traj', -1)  # same as above
read('O2Pt.traj', 0)   # first configuration
```
